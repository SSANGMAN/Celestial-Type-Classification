```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../dataset/train.csv', index_col = 0)
test = pd.read_csv('../dataset/test.csv', index_col = 0)
sample_submission = pd.read_csv('../dataset/sample_submission.csv', index_col = 0)
```

```python
tt = train.drop(columns = ['type', 'fiberID'], axis = 1)
tt.head()
```

```python
cols = train.columns[2:].tolist()
q1 = train.quantile(0.001)
q2 = train.quantile(0.999)
```

```python
idx = 0
for col in cols:
    x = q1[idx]
    y = q2[idx]
    qidx1 = train[train[col] < x][col].sort_values(ascending = False).index
    qidx2 = train[train[col] > y][col].sort_values().index
    a = 0
    b = 0.01
    for i1 in qidx1:
        train[col][i1] = x - np.random.uniform(a, b)
        a += 0.01
        b += 0.01
    a = 0
    b = 0.01
    for i2 in qidx2:
        train[col][i2] = y + np.random.uniform(a, b)
        a += 0.01
        b += 0.01
    idx += 1
train.head()
```

```python
d = 10 ** ((train['psfMag_u'] - train['petroMag_u'] + 5) / 5)
```

```python
def fib(x):
    if x <= 100:
        return 0
    elif x <= 200:
        return 1
    elif x <= 300:
        return 2
    elif x <= 400:
        return 3
    elif x <= 500:
        return 4
    elif x <= 600:
        return 5
    elif x <= 640:
        return 6
    else:
        return 7
    
train['fiberID2'] = train['fiberID'].map(fib)
test['fiberID2'] = test['fiberID'].map(fib)
train.head()
```

```python
for com in [train, test]:
    com['psf_mean'] = com[cols[:5]].mean(axis = 1)
    com['fiber_mean'] = com[cols[5:10]].mean(axis = 1)
    com['petro_mean'] = com[cols[10:15]].mean(axis = 1)
    com['model_mean'] = com[cols[15:]].mean(axis = 1)
    com['psf_median'] = com[cols[:5]].median(axis = 1)
    com['fiber_median'] = com[cols[5:10]].median(axis = 1)
    com['petro_median'] = com[cols[10:15]].median(axis = 1)
    com['model_median'] = com[cols[15:]].median(axis = 1)
    com['psf_std'] = com[cols[:5]].std(axis = 1)
    com['fiber_std'] = com[cols[5:10]].std(axis = 1)
    com['petro_std'] = com[cols[10:15]].std(axis = 1)
    com['model_std'] = com[cols[15:]].std(axis = 1)
    com['u_mean'] = com[cols[::5]].mean(axis = 1)
    com['g_mean'] = com[cols[1::5]].mean(axis = 1)
    com['r_mean'] = com[cols[2::5]].mean(axis = 1)
    com['i_mean'] = com[cols[3::5]].mean(axis = 1)
    com['z_mean'] = com[cols[4::5]].mean(axis = 1)
    com['u_median'] = com[cols[::5]].median(axis = 1)
    com['g_median'] = com[cols[1::5]].median(axis = 1)
    com['r_median'] = com[cols[2::5]].median(axis = 1)
    com['i_median'] = com[cols[3::5]].median(axis = 1)
    com['z_median'] = com[cols[4::5]].median(axis = 1)
    com['u_std'] = com[cols[::5]].std(axis = 1)
    com['g_std'] = com[cols[1::5]].std(axis = 1)
    com['r_std'] = com[cols[2::5]].std(axis = 1)
    com['i_std'] = com[cols[3::5]].std(axis = 1)
    com['z_std'] = com[cols[4::5]].std(axis = 1)
train.head()
```

```python
train['distance'] = 10 ** ((train['psf_mean'] - train['petro_mean'] + 5) / 5)
test['distance'] = 10 ** ((test['psf_mean'] - test['petro_mean'] + 5) / 5)
train.head()
```
